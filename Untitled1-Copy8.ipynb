```python
import matplotlib.pyplot as plt

round_thresh = 0.65

plt.figure()
height, width = img.shape
plt.imshow(img.astype(np.uint8), cmap='gray')

areas = []
alphas = []
arrayOfCenter = [] # mng to  tm ng trn
# print('Object      Area Perimeter Roundness')
for ind, contour in enumerate(contours):
    # tnh chu vi ca contour
    perimeter = cv2.arcLength(contour, True)
    # tnh din tch ca contour
    area = cv2.contourArea(contour)
    areas.append(area)
    
    # nu nh chu vi bng 0 hoc area nh hn threshold th ta t ngay gc alpha = 0
    if perimeter==0 or area<140:
        alpha = 0
    else:
        alpha = 4*np.pi*area/(perimeter**2)
        alphas.append(alpha)

    # V vng trn  ti tm ca contour tha mn iu kin l vng trn
    if alpha > round_thresh:
        moments = cv2.moments(contour)
        cx = int(moments['m10'] / moments['m00'])
        cy = int(moments['m01'] / moments['m00'])
        plt.plot(cx, cy, 'ro')
        arrayOfCenter.append([cx,cy])
```

```python
# Ly cc khung quanh ng trn
immm = img.astype(np.uint8)
for i in arrayOfCenter:
    print(i)
# Khung nh mi ng trn c khoanh
arrayOfFrame = []  
# Ly 40px cho mi khung
l = 20
for i in arrayOfCenter:
    a = []
    for y in range(i[1]-l, i[1]+l):
        row = []
        for x in range(i[0]-l, i[0]+l):
            row.append(immm[y][x])
        a.append(row)
    arrayOfFrame.append(a)
print(len(arrayOfFrame))
```

```python
# Hm ly contour ln nht
def find_max_contour(img):
    # Ly ra v tr cc contour (contour l 1 chui ta  (x_i, y_) cc im c cng  sng dc theo bin ca 1 object (t ang hiu l th))
    contours, _ = cv2.findContours(img.astype(np.uint8), # pylint: disable=unused-variable
                                  cv2.RETR_EXTERNAL,
                                  cv2.CHAIN_APPROX_NONE)
    num = 0
    maxContour = 0
    maxContourIndex = 0
    maxContourArray = []
    for index, cnt in enumerate(contours):
        num+= 1
        sizeOfContour = np.array(cnt).shape[0]
        if sizeOfContour > maxContour:
            x,y,w,h = cv2.boundingRect(cnt)
            maxContour = sizeOfContour
            maxContourIndex = index
            maxContourArray = cnt
    # V li contour ln nht
    newArray = np.zeros(img.shape)
    h = img.shape[0]
    w = img.shape[1]
    for i in range(0,h):
        for j in range(0, w):
            for k in maxContourArray:
                if k[0][0] == i and k[0][1] == j:
                    newArray[j][i] = 255
    return newArray
```

```python
# Hm xc nh xem b gch hay khng
def check_cross(img):
    # Ly contour ln nht
    maxContourArray = [] 
    maxContourArray = find_max_contour(img)
    # Fill contour ln nht
    from scipy.ndimage.morphology import binary_fill_holes
    clean_image = binary_fill_holes(maxContourArray)
    clean_image = np.float32(clean_image)
    # Opening  xo mt gch
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    opening = cv2.morphologyEx(clean_image, cv2.MORPH_OPEN, kernel)
    # Tr bn trc v sau opening  xc nh gch
    minus = clean_image - opening
    count = 0
    # m s lng pixel  xem c b gch khng, kh ln l c b gch
    h = img.shape[0]
    w = img.shape[1]
    for i in range(0,h):
        for j in range(0, w):
            if minus[j][i] > 0:
                count+=1
    print(count)
    # Chnh ngng nu cn
    if count  > 10:
        return True
    return False
```

```python
def get_circles_no_cross(centers):
    circleNoCross = []
    for i in centers:
        if check_cross(np.array(i)) == False:
            circleNoCross.append(i)
    return circleNoCross
```
