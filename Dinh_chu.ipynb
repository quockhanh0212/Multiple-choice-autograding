```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
c = cv2.imread("C_Dinh.png", 0)
plt.imshow(c, cmap='gray')
```

```python
c = cv2.adaptiveThreshold(c,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,31,7)
plt.imshow(c, cmap='gray')
```

```python
from scipy.ndimage.morphology import binary_fill_holes

# binary_fill_holes là 1 hàm binary delation
clean_image = binary_fill_holes(255-c)
clean_image = np.float32(clean_image)
plt.imshow(clean_image, cmap='gray')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (10,10))
opening = cv2.morphologyEx(clean_image, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap='gray')
```

```python
kernel = np.ones((10, 10), np.float32) 
image = cv2.erode(clean_image, kernel)  
plt.imshow(image, cmap='gray')
```

```python
tru = 255- (clean_image - image)
plt.imshow(tru, cmap='gray')
```

```python
rs1 = (255 -c)/255
rs1 = rs1.astype(np.float32)
plt.imshow(rs1, cmap='gray')
rs1[70:100,:30]
```

```python
rs2 = 255 - tru
plt.imshow(rs2, cmap='gray')
rs2[70:100,:30].astype(np.float32)
```

```python
rs3= abs(rs1 - rs2)
plt.imshow(rs3, cmap='gray')
rs3[70:100,:30]
```

```python
kernel = np.ones((3, 3), np.float32) 
rs4 = cv2.erode(rs3, kernel)  
plt.imshow(rs4, cmap='gray')
```
