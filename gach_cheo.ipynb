```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
```

```python
c = cv2.imread("c.png", 0)
plt.imshow(c, cmap='gray')
```

```python
d = cv2.imread("d.png", 0)
plt.imshow(d, cmap='gray')
```

```python
cross = cv2.imread("cross.png", 0)
plt.imshow(cross, cmap='gray')
```

```python
def count_contours(img):
    # Lấy ra vị trí các contour (contour là 1 chuỗi tọa độ (x_i, y_) các điểm có cùng độ sáng dọc theo biên của 1 object (t đang hiểu là thế))
    contours, _ = cv2.findContours(img.astype(np.uint8), # pylint: disable=unused-variable
                                  cv2.RETR_EXTERNAL,
                                  cv2.CHAIN_APPROX_NONE)
    num = 0

    # Vẽ contour
    obj_image = np.zeros(img.shape)
    for index, cnt in enumerate(contours):
        num+= 1
        cv2.drawContours(img, [cnt], 0, 255, -1)
    print(f'Number of contours: {num}')
```

```python
# khi vòng tròn khoanh dính vào chữ cái, phương pháp này sẽ không hiệu quả
count_contours(c)
```

```python
count_contours(d)
```

```python
count_contours(cross)
```
